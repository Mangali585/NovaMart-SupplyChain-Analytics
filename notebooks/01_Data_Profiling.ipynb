### Objective
Understand the dataset structure, identify data quality issues, and prepare the data for cleaning and business analysis.

```python
import pandas as pd

df = pd.read_csv(
    r"C:\Users\Mangali\NovaMart-SupplyChain-Analytics\data\raw\DataCoSupplyChainDataset.csv",
    encoding="latin1"
)

df.head()
```

```python
rows, cols = df.shape

print(f"Number of Rows    : {rows}")
print(f"Number of Columns : {cols}")
```

```python
df.info()
```

### Observations

- Dataset contains 180,520 records.
- Total columns: 53.
- Multiple numerical and categorical columns.
- Order Date and Shipping Date are stored as object (text).

```python
df.describe()
```

```python
df.describe(include="object")
```

```python
missing = df.isnull().sum()

missing[missing > 0].sort_values(ascending=False)
```

```python
missing_df = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Missing Percentage": (df.isnull().sum()/len(df))*100
})

missing_df = missing_df[missing_df["Missing Values"] > 0]

missing_df.sort_values("Missing Percentage", ascending=False)
```

```python
df.dtypes.value_counts()
```

```python
quality = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Missing %": round(df.isnull().mean()*100,2),
    "Data Type": df.dtypes
})

quality
```

```python
unique_values = pd.DataFrame({
    "Column": df.columns,
    "Unique Values": [df[col].nunique() for col in df.columns]
})

unique_values.sort_values("Unique Values")
```

```python
constant_cols = []

for col in df.columns:
    if df[col].nunique() == 1:
        constant_cols.append(col)

constant_cols
```

```python
duplicates = df.duplicated().sum()

print("Duplicate Rows :", duplicates)
```

# Exploratory Data Profiling
