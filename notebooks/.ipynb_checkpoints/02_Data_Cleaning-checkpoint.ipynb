# NovaMart Supply Chain Analytics

## Phase 3 : Data Cleaning & Feature Engineering

### Objective

Prepare the dataset for SQL analysis and Power BI dashboarding by
- Removing unnecessary columns
- Handling missing values
- Correcting data types
- Creating new business features

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "../data/raw/DataCoSupplyChainDataset.csv",
    encoding="latin1"
)
```

```python
df_clean = df.copy()
```

```python
columns_to_drop = [
    "Product Description",
    "Customer Email",
    "Customer Password",
    "Product Status",
    "Order Zipcode",
    "Order Item Cardprod Id",
    "Customer Zipcode"
]

df_clean.drop(columns=columns_to_drop, inplace=True)
```

```python
df_clean.shape
```

```python
missing = pd.DataFrame({
    "Missing": df_clean.isnull().sum(),
    "Percentage": round(df_clean.isnull().mean()*100,2)
})

missing[missing["Missing"] > 0]
```

```python
df_clean["Customer Lname"] = df_clean["Customer Lname"].fillna("Unknown")
```

```python
df_clean.isnull().sum().sort_values(ascending=False).head()
```

```python
print(df_clean["order date (DateOrders)"].head(10))
print(df_clean["shipping date (DateOrders)"].head(10))
```

```python
df_clean["order date (DateOrders)"] = pd.to_datetime(
    df_clean["order date (DateOrders)"]
)

df_clean["shipping date (DateOrders)"] = pd.to_datetime(
    df_clean["shipping date (DateOrders)"]
)
```

```python
print(df_clean["order date (DateOrders)"].min())
print(df_clean["order date (DateOrders)"].max())

print(df_clean["shipping date (DateOrders)"].min())
print(df_clean["shipping date (DateOrders)"].max())
```

```python
df_clean["Delivery Delay"] = (
    df_clean["Days for shipping (real)"] -
    df_clean["Days for shipment (scheduled)"]
)
```

```python
df_clean["Delivery Delay"].describe()
```

### Observation

The Delivery Delay feature represents the difference between actual shipping days and scheduled shipping days.

Interpretation:

• Negative → Delivered early

• Zero → Delivered on time

• Positive → Delivered late

This feature will be one of the core KPIs used in the Power BI dashboard.

```python
df_clean["Delivery Delay"].value_counts().sort_index()
```

```python
pd.crosstab(
    df_clean["Delivery Status"],
    df_clean["Late_delivery_risk"]
)
```

```python
df_clean["Benefit per order"].describe()
```

```python
(df_clean["Benefit per order"] < 0).sum()
```

```python
df_clean["Order Item Discount Rate"].describe()
```

```python
df_clean["Profit Margin"] = (
    df_clean["Benefit per order"] /
    df_clean["Sales"]
) * 100

df_clean["Profit Margin"] = df_clean["Profit Margin"].round(2)

df_clean["Profit Margin"].describe()
```

### Observation

Profit Margin helps compare profitability across products and markets irrespective of sales value.

This metric is more informative than absolute profit because it measures efficiency rather than volume.

# Save Cleaned Dataset

```python
df_clean.to_csv(
    "../data/cleaned/DataCoSupplyChain_Clean.csv",
    index=False
)

print("Cleaned dataset saved successfully!")
```
